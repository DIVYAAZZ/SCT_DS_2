```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

train_df = pd.read_csv('train.csv')

train_df['Age'].fillna(train_df['Age'].median(), inplace=True)
train_df.drop(columns='Cabin', inplace=True, errors='ignore')
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0], inplace=True)

sns.set(style="darkgrid")

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Titanic EDA Summary', fontsize=16)

sns.countplot(data=train_df, x='Survived', ax=axs[0, 0])
axs[0, 0].set_title('Survival Count')
axs[0, 0].set_xticklabels(['Not Survived', 'Survived'])

sns.countplot(data=train_df, x='Sex', hue='Survived', ax=axs[0, 1])
axs[0, 1].set_title('Survival by Gender')
axs[0, 1].legend(title='Survived', labels=['No', 'Yes'])

sns.countplot(data=train_df, x='Pclass', hue='Survived', ax=axs[1, 0])
axs[1, 0].set_title('Survival by Passenger Class')
axs[1, 0].legend(title='Survived', labels=['No', 'Yes'])

sns.histplot(data=train_df, x='Age', hue='Survived', bins=20, kde=True, ax=axs[1, 1], palette='Set1')
axs[1, 1].set_title('Age Distribution by Survival')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
